```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv ('data/original_sharktank_data.csv')
```

```python
deal_df = df[df.Deal != 'No']
deal_df

no_deal_df = df[df.Deal != 'Yes']
#no_deal_df
```

```python
deal_df
```

```python
industry_count = deal_df['Industry'].value_counts().tolist()
#industry_count
```

```python
industry_type = deal_df['Industry'].unique().tolist()
#industry_type
```

```python
plt.bar(industry_type, industry_count)
# add title and labels
plt.title('Which Industry Is Getting The Most Deals')
plt.xlabel('Industry')
plt.ylabel('# of Deals')
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', pad=15)

plt.savefig('Industry_Deals.png')
```

```python
no_industry_count = no_deal_df['Industry'].value_counts().tolist()
#industry_count

no_industry_type = no_deal_df['Industry'].unique().tolist()
#industry_type

plt.bar(no_industry_type, no_industry_count)
# add title and labels
plt.title('Which Industry Is Getting The Least Deals')
plt.xlabel('Industry')
plt.ylabel('# of No Deals')
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', pad=15)

plt.savefig('Industry_No_deals.png')
```

```python
gender_count = deal_df['Entrepreneur Gender'].value_counts().tolist()
#gender_count
```

```python
gender_type = deal_df['Entrepreneur Gender'].unique().tolist()
#industry_type
```

```python
plt.bar(gender_type, gender_count)
# add title and labels
plt.title('Which Gender Is Most Likely To Secure A Deal')
plt.xlabel('Gender')
plt.ylabel('# Of Each Gender')
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', pad=15)

plt.savefig('Deal_By_Gender.png')
```
