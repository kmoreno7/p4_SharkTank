```python
#define the X and y
X = df.drop(['Deal_Yes'], axis=1)
y = df['Deal_Yes']

for i, column in enumerate(X.columns):
  print(f"X[{i}] in .js informed by {column}")
```

```python
#Import Train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Split the data into X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.scale_
```

```python
#Get the mean from the scaler
from sklearn.datasets import make_classification
scaler.mean_
```

```python
#Get the var from the scaler (this will be squared in the javascript file to determine z score.)
scaler.var_
```

```python
# Train a Logistic Regression model and print the model score
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression().fit(X_train_scaled, y_train)
print('Model: Logistic Regression')
print(f'Training Score: {reg.score(X_train_scaled, y_train)}')
print(f'Testing Score: {reg.score(X_test_scaled, y_test)}')

```
