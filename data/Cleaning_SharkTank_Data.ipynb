```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['sharktank_data.csv']))
# Dataset is now stored in a Pandas Dataframe
```

```python
df
```

```python
# Drop the non-beneficial ID columns
drop_df = df.drop(columns=[
    'Season',
    'No. in series',
    'Episode',
    'Description',
    'State',
    'DEAL_Amount',
    'DEAL_Equity',
    'DEAL_Valuation',
    'Details / Notes',
    'Royalty Deal?',
    'Loan?',
    '$ per shark'
    ])

# Drop Sharks by column number
drop_df = drop_df.drop(drop_df.columns[7:15], axis=1)

drop_df
```

```python
#Drop NaN from the "ASK_Amount" column
drop_na_df = drop_df[drop_df['ASK_Amount'].notna()]
drop_na_df
```

```python
#Replace last NaN with 0 
fill_na_df = drop_na_df.fillna(0)
fill_na_df
```

```python
fill_na_df['ASK_Amount'] = fill_na_df['ASK_Amount'].str.replace(',', '')
fill_na_df['ASK_Amount'] = fill_na_df['ASK_Amount'].str.replace('$', '')

fill_na_df['ASK_Equity'] = fill_na_df['ASK_Equity'].str.replace('%', '')

fill_na_df['ASK_Valuation'] = fill_na_df['ASK_Valuation'].str.replace(',', '')
fill_na_df['ASK_Valuation'] = fill_na_df['ASK_Valuation'].str.replace('$', '')
```

```python
fill_na_df
```

```python
fill_na_df = fill_na_df.fillna(0)
```

```python
fill_na_df
```

```python
fill_na_df['ASK_Amount'] = fill_na_df['ASK_Amount'].astype(int)

fill_na_df['ASK_Equity'] = fill_na_df['ASK_Equity'].astype(float)
fill_na_df['ASK_Equity'] = fill_na_df['ASK_Equity'].astype(int)

fill_na_df['ASK_Valuation'] = fill_na_df['ASK_Valuation'].astype(int)
```

```python
fill_na_df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
clean_df = pd.get_dummies(fill_na_df,columns=['Deal','Industry','Entrepreneur Gender'])
clean_df = clean_df.drop(columns=['Deal_No'])
clean_df.head()
```
